```python
import numpy as np
import matplotlib.pyplot as plt
```

R(x)=max(0,x)
S(x)=1/(1+e^−x)
T(x)=(1−e−2x)/(1+e−2x)

```python
def R(x):
    return np.maximum(0,x) 
```

```python
x = np.arange(-4, 4, 0.01)
print(x)
R(x)
```

```python
def S(x):
    return 1/(1+np.exp(-x))
```

```python
S(x)
```

```python
def T(x):
    return (1-np.exp(-2*x))/(1+np.exp(-2*x))
```

```python
T(x)
```

```python
def U(x, t):
    return np.where(x>=t, 1, 0) 
```

```python
t=0
U(x,t)
```

```python
plt.plot(x,R(x),"r",label="Relu")
plt.plot(x,S(x),"g",label="Sigmoid")
plt.plot(x,T(x),"b",label="Tanh")
plt.plot(x,U(x,0),"y",label="Treshold-Funktion")
plt.legend()
plt.show()
```
