```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))
```

```python
def ReLU(x):
    return np.maximum(x,0)
```

```python
def Threshold0(x):
    return np.where(x < 0, 0, 1)
```

```python
def Identity(x):
    return x
```

```python
data = np.arange(0,10)
data
```

```python
y2 = 1.0 * (data > 6)
y2
```

```python
eta = 0.001
eta
```

```python
epochs = 100
epochs
```

```python
# w = np.full(10, 0)
w = 0.5
w
```

```python
# b = np.full(10,0)
b = -2
b
```

```python
for epoch in range(epochs):
    
    print("epoch {0}".format(epoch))
    
    for i, x in enumerate(data):
#        print(str(i) + " " + str(x))

        #        y = Sigmoid(w * x + b)
        y = ReLU(w * x + b)
#        print("\t{0}".format(y))

        error = (y-y2[i])**2
#        print(error)
        
        dw = -eta * (y-y2[i])
        # print(dw)

        w += dw
        # print(w)

        db = -eta * np.sign(y-y2[i])
        # print(db)

        b += db
        # print(b)

        print("x {0}\ty {1}\tE {2:.3f}".format(x, y, error))
        print("\tw\t{0:.3f}\t{1:.3f}".format(w, dw))
        print("\tb\t{0:.3f}\t{1:.3f}".format(b, db))

    y3 = Threshold0(w * data + b)
    print(y3)
```
